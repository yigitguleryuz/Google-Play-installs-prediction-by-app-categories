## DATA PREPROCESSING

```python
#I'll delete unimportant columms
df = df.drop("App", axis=1)
df = df.drop("Last Updated", axis = 1)
df = df.drop("Current Ver", axis = 1)
df = df.drop("Genres", axis = 1)
```

```python
# Lets deal with dtyeps.
df.info()
```

```python
df[df["Rating"].isnull()]
#I will delete the null ratings because i dont want to deal with summarising.
df = df.dropna(subset=["Rating"])
```

```python
#Lets deal with other nulls
df = df.dropna(subset=["Android Ver"])
```

```python
df.info()
```

```python
df["Size"].unique()
```

# Dtype Conversion for integer supposed to be

```python
print(df["Rating"].unique())
print("-" * 30)  
print(df["Reviews"].unique())
print("-" * 30)  
print(df["Size"].unique())
print("-" * 30)  
print(df["Installs"].unique())
print("-" * 30) 
print(df["Price"].unique())
print("-" * 30)
print(df["Content Rating"].unique())
print("-" * 30)
print(df["Android Ver"].unique())
print("-" * 30)

#Rating
df["Rating"] = df["Rating"].astype(float)

#Reviews
df["Reviews"] = df["Reviews"].astype(float)

#Size
m_mask = df["Size"].str.contains("M", na=False)
df.loc[m_mask, "Size"] = (
    df.loc[m_mask, "Size"]
    .str.replace("M", "", regex=False)
    .astype(float) * 1024
).astype(str)

k_mask = df["Size"].str.contains("k", na=False)
df.loc[k_mask, "Size"] = (
    df.loc[k_mask, "Size"]
    .str.replace("k", "", regex=False)
).astype(str)
df["Size"] = df["Size"].replace("Varies with device", np.nan)
df["Size"] = pd.to_numeric(df["Size"], errors='coerce')
df["Size"].fillna(df["Size"].mean(), inplace=True)

#Installs
df["Installs"] = df["Installs"].str.replace("+", "", regex=False)
df["Installs"] = df["Installs"].str.replace(",", "", regex=False)
df["Installs"] = df["Installs"].astype(int)

#Price
df["Price"] = df["Price"].str.replace("$", "", regex=False)
df["Price"] = df["Price"].astype(float)

#Android Ver
df["Android Ver"] = df["Android Ver"].fillna(df["Android Ver"].mode())

df.loc[df["Android Ver"] == "Varies with device", "Android Ver"] = np.nan

df["Android Ver"] = df["Android Ver"].astype(str).str.extract(r'^(\d+\.\d+)')
df["Android Ver"] = df["Android Ver"].astype(float)
df["Android Ver"].fillna(df["Android Ver"].mean(), inplace=True)
```
